```python

# Tải dữ liệu
fifa_2022 = pd.read_csv('FIFA22_official_data[1].csv', encoding='latin-1')
fifa_2023 = pd.read_csv('FIFA23_official_data[1].csv', encoding='latin-1')

# Kết hợp dữ liệu
data = pd.concat([fifa_2022, fifa_2023], ignore_index=True)

# Mô tả dữ liệu
print(data.describe())

# Biểu đồ phân bố chiều cao cầu thủ
plt.figure(figsize=(10, 6))
data['Height'] = data['Height'].str.rstrip('cm').astype(float)
plt.hist(data['Height'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Phân bố chiều cao cầu thủ')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Số lượng')
plt.show()

# Biểu đồ phân bố mức lương cầu thủ
#data['Wage'] = data['Wage'].str.replace('€', '').str.replace('K', '').astype(float) * 1000
plt.figure(figsize=(10, 6))
plt.hist(data['Wage'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Phân bố mức lương cầu thủ')
plt.xlabel('Mức lương (€)')
plt.ylabel('Số lượng')
plt.show()

# Biểu đồ phân bố kỹ năng di chuyển
plt.figure(figsize=(10, 6))
plt.hist(data['Skill Moves'].dropna(), bins=5, edgecolor='k', alpha=0.7)
plt.title('Phân bố kỹ năng di chuyển cầu thủ')
plt.xlabel('Kỹ năng di chuyển')
plt.ylabel('Số lượng')
plt.show()
```

Tiền xử lý dữ liệu

```python
# Xử lý giá trị thiếu
data['Height'].fillna(data['Height'].median(), inplace=True)
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Skill Moves'].fillna(data['Skill Moves'].median(), inplace=True)
data['Overall'].fillna(data['Overall'].median(), inplace=True)
data['Potential'].fillna(data['Potential'].median(), inplace=True)

# Chuyển đổi dữ liệu mục tiêu
data['Transfer_Status'] = data['Transfer_Status'].map({'Transfer': 1, 'No Transfer': 0})


```

Sử dụng kỹ thuật như SMOTE (Synthetic Minority Over-sampling Technique) nếu dữ liệu bị mất cân bằng.

```python

X = data[['Overall', 'Potential', 'Skill Moves', 'Height', 'Age']]
y = data['Transfer_Status']

# Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

# Tạo mô hình Neural Network
def create_nn_model(activation='relu', optimizer='adam'):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation=activation))
    model.add(Dense(32, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Huấn luyện mô hình với Early Stopping
model = create_nn_model()
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# Đánh giá mô hình
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

# Tạo mô hình với các tham số khác nhau
def create_nn_model(activation='relu', optimizer='adam'):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation=activation))
    model.add(Dense(32, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Định nghĩa hàm để chạy GridSearchCV
def build_model(activation, optimizer):
    model = create_nn_model(activation, optimizer)
    return model

# Định nghĩa các tham số tìm kiếm
param_grid = {
    'activation': ['relu', 'tanh'],
    'optimizer': ['adam', 'sgd']
}

# Grid Search cho Neural Network
best_model = None
best_score = 0

for activation in param_grid['activation']:
    for optimizer in param_grid['optimizer']:
        model = build_model(activation, optimizer)
        history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)
        score = model.evaluate(X_test, y_test, verbose=0)[1]  # accuracy
        if score > best_score:
            best_score = score
            best_model = model

print(f"Best accuracy: {best_score}")
```
